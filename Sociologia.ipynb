# MARX  
---

### **Prólogo a la “Contribución a la crítica de la Economía política” (1847)**

- Las condiciones jurídicas y políticas no pueden comprenderse por sí mismas. Estas radican en las condiciones materiales de vida. Hay que buscar la anatomía de la sociedad civil en la economía política.
- En la producción social de su existencia, los hombres establecen relaciones necesarias e independientes de su voluntad. Relaciones de producción, correspondientes a un determinado estadio evolutivo de sus fuerzas materiales productivas. Estas relaciones de producción constituyen la estructura económica de la sociedad, siendo esta la base real sobre la cual se alza el edificio jurídico y político.
- A estas estructuras económicas les corresponden determinadas formas de conciencia social. El modo de producción determina el proceso social, político e intelectual de la vida en general. La existencia y lugar social que ocupan los hombres determinan su conciencia.
- En un punto de su existencia, las fuerzas materiales productivas entran en conflicto con las relaciones de producción, iniciando así una época de revolución social.
- Esta nueva formación social procederá hasta que no se hayan desarrollado las fuerzas materiales productivas para las que sean ampliamente suficientes. Jamás ocuparán su lugar nuevas y superiores relaciones de producción antes de que las condiciones de su existencia se incuben en el propio seno de la antigua sociedad.
- Las relaciones de producción burguesas modernas son las últimas formas antagónicas del proceso social de la producción. Con esta forma social concluye la prehistoria de la sociedad humana.

---

### **Manifiesto comunista (1848)**

- La historia de todas las sociedades es la historia de una lucha de clases, entre opresores y oprimidos. Esta lucha termina con la transformación revolucionaria de toda la sociedad o con el hundimiento de las clases en lucha.
- Hoy en día, la sociedad moderna burguesa ha simplificado la disputa entre dos grandes clases: burguesía y proletariado.

**La burguesía:**
1. Ha disuelto la dignidad humana en el valor de cambio sin escrúpulos, y no ha dejado otro lazo entre los hombres que el desnudo interés.
2. Obliga a todas las naciones a apropiarse de su modo de producción si no quieren sucumbir.
3. Ha aglomerado a la población, concentrando la propiedad, lo que ha resultado en la centralización del poder político.

**En la misma medida que se desarrolla la burguesía, se desarrolla el proletariado:**
- Son los obreros que tienen que venderse al por menor como una mercancía cualquiera, igualmente expuestos a las vicisitudes de su propia concurrencia y las oscilaciones del mercado. La propia dinámica del sistema hace que todas las clases intermedias caigan en el proletariado. Estos proletarios son los hombres que portarán las armas para sublevarse ante la burguesía, sus explotadores.

**Conclusión:**  
El proletariado carece de propiedad y solo puede conquistar las fuerzas productivas sociales aboliendo su propio modo de apropiación. Con ello, caería el modo de apropiación existente hasta ahora.

### **Trabajo asalariado y capital (1849)**

En la actualidad, se vive bajo el dominio de la producción capitalista. Una clase numerosa y cada vez más extensa de la población solo puede existir trabajando a cambio de un salario para los propietarios de la producción. La fuerza de trabajo es la propia actividad vital del obrero, que debe venderla al capitalista para asegurarse los medios de vida necesarios.

La fuerza de trabajo es, en la actual economía capitalista, una mercancía con la especial virtud de ser creadora de valor. Si se sabe emplear, puede generar mayor valor del que ella misma cuesta. Este valor añadido es captado por los propietarios de los medios de producción, y los trabajadores no son retribuidos con la totalidad del valor que crean con su fuerza de trabajo. Con cada avance tecnológico y científico, crece este remanente de su producción diaria sobre sus costes diarios, reduciendo así cada vez más la parte en que el obrero produce el equivalente a su salario.

**La clase obrera** es la que produce todos los valores, ya que el valor no es más que otro término para expresar el trabajo. Sin embargo, estos valores producidos no les pertenecen, sino a los propietarios de la fuerza de trabajo de los obreros. Hay una contradicción entre capital y trabajo.

Esto acaba por crear un conflicto en el cual forzosamente debe perecer la actual economía capitalista. Existe una división en la sociedad: por un lado, una reducida clase fabulosamente rica; por otro, una inmensa clase de asalariados que no tienen ninguna garantía ante la penuria extrema.

**Es posible un nuevo orden social** en el que desaparezcan estas clases. Mediante el aprovechamiento y el desarrollo de las inmensas fuerzas productivas ya existentes de todos los individuos, e imponiendo el deber general de trabajar, se dispondrá para todos, en proporciones cada vez mayores, de los medios necesarios para vivir, disfrutar la vida, y para educar y ejercer todas las facultades físicas y espirituales.

---

### **Burguesía**

**Burguesía industrial:**  Propietarios de las fábricas e industrias. Promueven políticas de libre comercio e industrialización, expansión del mercado, reducción de costos laborales y materias primas baratas.

**Burguesía financiera:**  Banqueros, prestamistas. Se centran en la estabilidad del sistema financiero y en la especulación. Sus intereses pueden ser contrarios a los de la burguesía industrial porque no tienen en cuenta a la producción.

**Burguesía comercial:**  Comerciantes y empresarios enfocados en el comercio y distribución de mercancías. Enfocados en la apertura de los mercados.

### **Pequeña burguesía**

**Artesanos y pequeños comerciantes:**  Propietarios de pequeñas empresas y talleres. Interesados en el proteccionismo y la regulación de la competencia con las grandes empresas, así como en la estabilidad económica.

**Profesionales y funcionarios:**  Comprende a profesionales liberales y funcionarios públicos de mediana jerarquía. Sus intereses son mantener su estatus social y económico a través de políticas que aseguren sus profesiones.

### **Terratenientes y Aristocracia**

Su poder disminuyó con el auge de la burguesía. Su interés es preservar sus tierras y privilegios tradicionales. Suelen alinearse con otras clases dominantes para mantener su estatus.

---

### **Proletariado**

Son los trabajadores asalariados, cuya actividad vital es el trabajo que venden como mercancía para los capitalistas.

- **Prole industrial:** Trabajadores en fábricas, interesados en salarios más altos y mejores condiciones laborales.
- **Prole agrícola:** Trabajadores rurales en tierras de grandes propietarios, interesados en la reforma agraria.

### **Campesinado**

- **Pequeños propietarios:**  Campesinos parcelarios dueños de sus pequeñas tierras, las cuales trabajan ellos mismos. Están interesados en la protección de sus propiedades.

- **Trabajadores sin tierras:**  Campesinos que trabajan para los pequeños propietarios, interesados en la obtención de tierras.

### **Lumpenproletariado**

Individuos marginados y desempleados, a veces involucrados en actividades ilegales. Carecen de conciencia de clase y a menudo son utilizados por otras clases para sus propios fines políticos.

# MAX WEBER
---

**Weber es un sociólogo antipositivista.** Rechaza el uso de los métodos de ciencias físico-naturales en las ciencias sociales. Para Weber, la sociología es la ciencia que estudia la acción social para explicarla causalmente en su desarrollo y efectos.

**Sociedad:** Es un conjunto de relaciones sociales entre individuos, es decir, un entramado de interacciones entre individuos que actúan.

**Acción social:** Es la acción de un individuo que toma en consideración las conductas de otros y se orienta en consecuencia.

Tipos de acción social según Weber:
1. **Acción racional con arreglo a fines:** Propósito claro y consciente. Se evalúan los medios adecuados para conseguir un fin específico. La racionalidad es subjetiva.
2. **Acción racional con arreglo a valores:** Motivada por convicciones o ideas. Hay conciencia y voluntad de asumir los riesgos.
3. **Acción afectiva:** Motivada por estados emocionales o sentimientos.
4. **Acción tradicional:** Guiada por costumbres o hábitos muy arraigados.

**Relación social:** Es el vínculo y marco en el cual los actores se comportan de forma conjunta.

---

### Ética protestante y espíritu del capitalismo

Es un estudio sobre cómo determinados pensamientos y mecanismos religiosos intervinieron y desarrollaron la mentalidad capitalista. No trabaja con leyes causales, sino que plantea dos entes ideales: **espíritu capitalista** y **ética protestante**, a los cuales encuentra afinidades electivas. Estas afinidades son aspectos que se complementan y contribuyen al desarrollo de los entes ideales.

**Espíritu del capitalismo:** Es un sistema moral que Weber desarrolla en base a las máximas de Franklin. Este sistema normativo establece como fin último de todo hombre honrado la obtención de la mayor ganancia posible. Incluye conductas racionales a la hora de perseguir estos beneficios.

**Máximas de Franklin:**
1. El tiempo es dinero.
2. El crédito es dinero.
3. El dinero es fértil y reproductivo.
4. El que gasta inútilmente un céntimo diario derrocha seis libras en un año, que constituyen el precio de usar 100.

**Ética protestante:** Para Weber, la ética protestante, o más específicamente la doctrina calvinista, impulsada por su particular teoría de la predestinación, promovía una mentalidad de trabajo duro, sacrificio y acumulación de riqueza como signos de la gracia divina. La riqueza material era vista como el tesoro de Dios en la tierra.

La **predestinación** es la idea de que Dios ya ha decidido quiénes serán salvados. Esto generaba ansiedad entre los creyentes por saber si formaban parte de los elegidos. La forma de conocer su destino se podía evidenciar a través de la dedicación en una profesión y el trabajo incesante.

**Resultados:**
1. **Ascetismo:** La moderación se consideraba una virtud religiosa.
2. **Vocación:** Cada individuo estaba llamado a un trabajo específico en la vida. Seguir fervientemente esta profesión era la forma de servir a Dios.
3. **Acumulación de capital:** Se resistían a derrochar dinero en placeres terrenales. En su lugar, lo invertían en sus trabajos o empresas.

**Conclusión:** La combinación de estas características dio lugar a la propagación del capitalismo racional moderno, caracterizado por una moderación racional del dinero, una regulación precisa de la vida social, la extensión del cálculo exacto en la economía y la aplicación de métodos científicos en la producción.

# DURKHEIM
---
Durkheim busca hacer de la sociología una ciencia. Para ello, delimita su **objeto de estudio** y su **método**.

### Objeto: Hecho social
Un hecho social es toda manera de hacer, pensar o sentir, susceptible de ejercer sobre el individuo una coacción exterior. Es general en una sociedad determinada y tiene una existencia propia independiente de sus manifestaciones individuales, es decir, son fenómenos sociales. Los hechos sociales tienen tres características:

1. **Exterioridad**: Existen por fuera de la conciencia de los individuos.
2. **Generalidad**: Son parte de la sociedad en su conjunto, comunes a todos los miembros.
3. **Coactivos**: Ejerce una presión sobre el comportamiento de los individuos.

### Método: Causal explicativo – Positivismo
1. **Hechos sociales como cosas**: Hay que tratar los hechos sociales como cosas para poder analizarlos de manera objetiva. Esto implica realizar un análisis, estudio y comparación con otros hechos de cualidades similares.
2. **Descartar nociones subjetivas previas**.
3. **Aislar los hechos sociales** de sus manifestaciones individuales, lo cual se hace a través de la estadística.
---
### Anomia
La **anomia** es una situación de falta de solidaridad social. No existen normas morales, lo que conlleva a que no se reconozcan ni se respeten las leyes. Esto significa una desintegración social.

### Forma de salvar la anomia
1. Crear **instituciones** con autoridad que reglamenten a la sociedad.
2. Aplicar el **derecho positivo**.
3. Implementar medidas a través de **instituciones educativas**.
---
### Sociedad primitiva
Existe **homogeneidad** y ausencia de individualidad.
- **Solidaridad mecánica**: Es el vínculo social basado en la semejanza de sus miembros.
- **Derecho represivo**: Debido a las semejanzas entre los miembros, toda infracción se considera contra el conjunto social. Hay una fuerte manifestación de la conciencia colectiva y se reprimen y castigan las conductas indebidas.

### Sociedad moderna
Hay una **diferenciación** entre los miembros; existen individualmente.
- **Solidaridad orgánica**: El vínculo social surge de la interdependencia de las especializaciones de los individuos.
- **Derecho restitutivo**: Debido a la heterogeneidad, las infracciones son individuales. Se crean obligaciones de reponer los daños causados.
---
### División del trabajo social
La **división del trabajo social** es la diferenciación y especialización de roles y tareas en una sociedad. Esta división es la fuente de la **solidaridad orgánica** y de la formación de individuos con conciencia propia, siendo la base de las sociedades modernas.

# Polanyi
---
**Según Polanyi, la civilización se sostenía sobre cuatro instituciones principales:**

1. **Sistema de equilibrio entre potencias** (Política)  
2. **Estado liberal** (Política)  
3. **Patrón oro internacional** (Económico)  
4. **Método autorregulador** (Económico)

Polanyi creía que la sociedad del siglo XIX se derrumbó porque estaba fundada sobre instituciones representativas de ideas utópicas que no podían existir sin aniquilar la sustancia humana y la naturaleza de la sociedad.

**Sobre la paz en el siglo XIX, según Polanyi:**

- La paz se mantuvo porque primero la **Santa Alianza** y después el **Concierto Europeo** se dedicaron a preservarla. Estas alianzas actuaban motivadas por intereses patrimoniales: si una guerra entre potencias se desataba, sus riquezas peligraban.

1. **Santa Alianza:** Actuaba principalmente suprimiendo de manera belicosa los movimientos nacionalistas en su génesis.
2. **Concierto Europeo:** Utilizaba las altas finanzas para influir. Para esta época, el interés por la paz ya se había asentado en el seno de las naciones, y la costumbre de invertir se había consolidado. Esto permitió que las organizaciones financieras internacionales se inmiscuyeran en la diplomacia entre potencias. Manipulaban a través del dinero y el crédito, manteniendo a raya a las potencias menores que pretendieran alterar el status quo.

**Fracaso del orden internacional:**

Este orden internacional fracasó cuando **Alemania** decidió aliarse con **Italia** y **Austria-Hungría**. Algunas décadas después, se formó una contraalianza encabezada por **Gran Bretaña**, que en aquellos años había sido la principal promotora de la paz. Ahora existían dos bloques antagonistas y no había un tercer grupo que pudiera unirse a los otros para equilibrar al grupo que intentara aumentar su poder.

La incoordinación entre potencias se acentuó, y la desintegración económica dejó sin capacidad operativa a las altas finanzas. La guerra no tardó en llegar.

# Bourdieu
---

### Las estructuras sociales de la economía

**Economía como hecho social total:** Bourdieu destaca la importancia de considerar las prácticas económicas como hechos sociales, lo que implica entenderlas dentro de un entramado social. Esto contrasta con la visión clásica económica que las considera desde una abstracción, separadas del contexto social.

---

### Conceptos fundamentales:

**1. Campo:**

Es el espacio social donde los agentes compiten por el capital. Cada campo tiene sus propias normas y jerarquías. Dentro de ese campo, se compite por el capital valorado en ese contexto.

**2. Capital:**

Son los recursos y poderes que los agentes acumulan y utilizan para mejorar su posición dentro del campo.

- **Capital económico:** Recursos financieros y materiales.
- **Capital social:** Redes y conexiones sociales.
- **Capital cultural:** Conocimiento, habilidades, educación y otras cualidades culturales.
- **Capital simbólico:** Representa el prestigio, honor y reconocimiento que una persona recibe en la sociedad. Se basa en la percepción social de los otros tipos de capital.

**3. Hábitus:**

Se refiere a las disposiciones que guían las prácticas de los individuos. El hábitus es adquirido a partir de las condiciones de existencia en una clase social y también es construido por la experiencia. Tiende a garantizar la conformidad de las prácticas a lo largo del tiempo.

- El hábitus permite la libre producción de pensamientos, percepciones y acciones dentro de los límites de su propia producción. Asegura una libertad condicionada entre el determinismo y la creación imprevisible.
- El hábitus es influenciado por la estructura social, pero a la vez colabora en la construcción y cambio de ésta. La fidelidad al hábitus es inconsciente.

**4. Ortodoxia:** 

Es el conjunto de agentes que tienen más poder y se ubican en la cima de la escala jerárquica de su campo. Se resisten al cambio imponiendo sus normas.

**5.Heterodoxia:** 

Son los agentes con menos capital y poder dentro del campo. Luchan por el cambio, buscando transformarse en los dominantes.

# GRANOVETTER
---
Lo que pretende el autor es darle la importancia necesaria a las relaciones sociales al analizar la acción económica de los individuos. A esto le llama incrustación. 

**Introducción al problema de la incrustación**
- **Incrustación:** Es el argumento de que la conducta y las instituciones están sumamente constriñidas por las relaciones sociales. 
Granovetter postula que el nivel de incrustación en las políticas sociales premercantiles era más bajo de lo que los sustantivistas proclamaban y que, con el paso a la modernización, ha cambiado menos de lo que creen. Además, aclara que ese nivel fue y será siempre mayor de lo que proclaman los formalistas.

---

**Concepción de la acción humana en la sociología y la economía**

- **Concepción sobresocializada:** Considera a las personas inmensamente sensibles a la opinión de los demás. Por lo tanto, estas son completamente obedientes a los dictámenes de los sistemas consensualmente desarrollados de normas y valores. Esta obediencia se internaliza a través de la socialización, por ende, no se percibe como carga.

- **Concepción infrasocializada:** Rechaza cualquier influencia de la estructura o relaciones sociales en las conductas. Las personas son seres atomizados y portadores de información perfecta sin contacto prolongado con otros. Esta noción sobrevive porque al eliminar de la ecuación las relaciones sociales también se eliminan conjuntamente las fricciones y problemas del orden caracterizados por la deshonestidad y la falta de confianza.

Irónicamente, estas dos concepciones opuestas operan de la misma manera. Ambas tienen en común al actor como un ente atomizado. La forma en que funcionan es bastante mecánica. Una vez que se sabe la clase social o qué es lo que le reporta más utilidad al actor, ya se sabe de forma automática cómo va a actuar.
En realidad, el ser humano no es esclavo del guión asignado a su rol social ni es un átomo aislado del contexto. Éste actúa en un plano que lo influencia y, a su vez, él participa en su construcción.

---

**Incrustación, falta de honestidad y confianza en la vida económica**

- **Oportunismo:** Interés propio perseguido con astucia, donde los agentes capaces de ocultar sus motivos consiguen ventajas transaccionales.

  - **Solución infrasocializada al oportunismo:** Existen dispositivos sociales e instituciones que evolucionaron para resolver los problemas económicos. La falta de honestidad se evita porque estos dispositivos se evolucionaron para hacerla muy costosa. La confianza no entra en juego, ya que estos dispositivos la reemplazan. 

    - **Problemas:** Ignora el grado en que las relaciones sociales evitan la deshonestidad. Pensar que los dispositivos sociales sustituyen a la confianza hace pensar que cualquier individuo racional tiene motivos para desarrollar formas de evadirlos.

  - **Solución sobresocializada al oportunismo:** Moralidad generalizada, los individuos evolucionaron hacia acuerdos implícitos sobre cierto tipo de respeto, acuerdos que son esenciales para la supervivencia.

     - **Problemas:** Recurre a una respuesta automática y generalizada; además, no tiene en cuenta la eficacia de esta moralidad cuando la acción implica costes muy elevados.

**Incrustación:** Acentúa el papel de las relaciones personales concretas y las estructuras de esas relaciones para generar confianza y honestidad. Las relaciones económicas continuas suelen llegar a cargarse de un contenido social que configura confianza y ausencia de oportunismo. Esta postura no niega la moralidad generalizada ni los dispositivos sociales, pero las ubica en un segundo orden. En la práctica, aceptamos estas vías una vez no hay a nuestro alcance información de primera mano. Hay que advertir que, al igual que las relaciones sociales mitigan el oportunismo, también aumentan el beneficio que ésta reporta. Por ende, no se pretenden hacer predicciones de gran alcance sobre el orden o el desorden.

---

**El problema de los mercados y las jerarquías**

Según Williamson, las transacciones económicas inciertas y recurrentes requieren inversiones específicas y suelen darse dentro de empresas jerárquicas. Esto sucede por dos razones: Primero, la capacidad limitada de proveer contingencias. Segundo, por la mitigación del oportunismo que implican las relaciones de autoridad. 
Granovetter opina que esto es errado. Cree que Williamson subestima la incrustación de las redes sociales entre empresas. La visión que tiene Williamson sobre el mercado es infrasocializada y la que tiene sobre las estructuras jerárquicas es sobresocializada. El autor sostiene que hay mucha evidencia de la superposición entre relaciones sociales y las transacciones entre empresas. Esto niega que las operaciones de mercado se acerquen al funcionamiento del estado de naturaleza joveciano que pueda resolverse con la internalización de una estructura jerárquica.

**Discusión**

Como los argumentos de la elección racional están construidos exclusivamente para individuos e intereses económicos, no están en consonancia con la posición de la incrustación. Sin embargo, el supuesto de la acción social puede ser muy útil si esta racionalidad la estudiamos teniendo en cuenta la incrustación. Esto se puede percibir dándonos cuenta de que no sólo existen metas económicas, sino también de sociabilidad, aprobación, estatus o poder. 
Sin una correcta comprensión de las redes de relaciones sociales no se puede establecer una conexión entre las teorías de nivel micro y macro.

# GINO GERMANI

---

## **Transición de Sociedad Tradicional (ST) a Sociedad Industrial (SI)**

- Surgimiento de una sociedad radicalmente distinta, abarca a todo el mundo y a todos los aspectos de la vida.  
- Implica un proceso que es vivido como crisis, ya que es una ruptura con el pasado.

### **Cambio en los fundamentos de la estructura**  
#### **Tipos de acción social:**  
  - **Acción prescriptiva:** Las personas siguen un patrón relativamente fijo, lo que se explicita en la forma de sentir y de actuar. Las variabilidades del comportamiento están alrededor de una pauta moral. **ST** 
  - **Acción electiva:** Las acciones humanas se realizan a base de elección; no por ello están menos reguladas. No se prescribe un comportamiento, sino que es la forma de elegir el comportamiento lo prescripto. **SI**

#### **Actitudes frente al cambio**  
- **Pautas preestablecidas:** Lo nuevo es rechazado; el cambio es anormal y constituye siempre una violación de las normas. **ST**
- **Cambio en la sociedad industrial:** El cambio se toma como un fenómeno social previsto e instituido en las normas. **SI**

#### **Especialización de instituciones**  
- **Estructura diferenciada:** La estructura tiene pocas funciones, y las instituciones abarcan la mayoría de las actividades humanas. **ST** 
- **Estructura especializada:** Las distintas funciones se diferencian cada vez más. La especialización de esferas normativas genera también una pluralidad de sistemas valorativos. **SI**

### **Cambio en los tipos de relaciones sociales**  
Hay una oposición de grupos primarios y secundarios. Pasaje de relaciones personales a impersonales. Se acentúan las relaciones secundarias orientadas a la eficiencia, aunque no se eliminan las primarias, ya que son necesarias para la integración normativa.  
- **Afectividad:** La gratificación de la acción social es inmediata y directa. **ST**
- **Neutralidad afectiva:** La expresión del afecto es postergada; se concibe a la acción como instrumental para un bien futuro. **SI**

- **Particularismo:** Las expectativas dirigidas a un rol determinado están dirigidas a una persona en concreto. **ST** 
- **Universalismo:** Las expectativas de los roles están dirigidas a cualquiera de una categoría de personas definida. **SI** 

- **Difusión:** No están definidos los límites, las responsabilidades ni las obligaciones. **ST**  
- **Especificidad:** Las relaciones son muy especializadas y concretas. **SI** 

- **Adcripción:** Los roles son adjudicados en función de lo que es. **ST**  
- **Desempeño:** El rol es adjudicado en función de lo que se logra hacer. **SI**

- Orientación hacia intereses privados **ST** 
- Orientación hacia intereses colectivos **SI**
### **Cambios en los tipos de personalidad:**  
- **Personalidad tradicional:** Estructura mental dominada por la internalización de normas prescriptas. **ST**
- **Personalidad giroscopio:** Etapa temprana de transición; sigue a través de decisiones propias y ausencia de cursos prefijados hacia metas internalizadas. **SI**  
- **Personalidad radar:** Persiste un carácter electivo en la acción, pero orientado a una sensibilidad por las acciones; le da más importancia a la necesidad de guiarse por los demás. **SI**

---

## **Condiciones e implicancias del desarrollo**  

### **Estratificación social:**  
  - **Condiciones:** La división del trabajo debe modificar los oficios, generando nuevos sectores. En la certificación, los roles deben transformarse en adquiridos, asegurando una mayor movilidad social. **ST**  
  - **Implicancia:** Mayor preponderancia de los estratos inferiores, tendientes a pagar hasta sus últimas consecuencias; bienes ideales igualitarios. **SI**  
### **Organización política:**  
  - **Condición:** Organización racional del estado y lealtad al estado nacional. **ST** 
  - **Implicancia:** Ruptura y traslado desde las comunidades locales hacia las nacionales; mayor participación de los estratos populares en la dirección del estado. **SI**
### **Organización familiar:**  
  - **Condición:** Restricción de los lazos familiares primarios, suprimiendo las relaciones difusas a su mínima expresión. **ST**   
  - **Implicancia:** El parentesco se reduce hasta la familia nuclear aislada. **SI**

## **Factores que intervienen en el proceso de transición**  
- Se debe efectuar una separación del conocimiento científico de la naturaleza respecto de toda actividad intelectual o forma de conocimiento ajeno (teología, filosofía, etc.) 
- **Técnica:** Todos los procedimientos de producción y distribución vinculados a la religión y tradición deben desaparecer.  
- **Economía:** Diferenciación de las instituciones específicamente económicas orientadas hacia el requisito de eficiencia y cuyos principios básicos son la racionalidad instrumental y la institucionalización del cambio.  
- **Conocimiento:** La secularización debe extenderse a todos los niveles del conocimiento, desde lo normativo y lo motivacional, esto es, desde las instituciones y de las propias personalidades de los individuos.

## **Asincronía**  
- Se refiere a la disparidad en el grado de modernización en diferentes sectores de la sociedad. Distintos sectores de la estructura social se encuentran en etapas diferentes del desarrollo, pudiendo ser:  
  - **Demográfica**  
  - **Geográfica**  
  - **Institucional**  
  - **Motivacional**, etc.

### **Efectos de la sincronía**  
- **Efecto demostración:** Impulsa a favor de la modernización de los sectores atrasados cuando estos sectores tradicionales y menos desarrollados toman conciencia del nivel y capacidad de consumo de sectores más avanzados.  
- **Efecto fusión:** Aprovechamiento por parte de sectores tradicionales de algunos aspectos de la modernización, preservando esos rasgos tradicionales.
